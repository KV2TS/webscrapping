```python
# Fonction pour retourner tous les codes IATA associés à une ville
def get_all_iata(city_name, df):
    results = df[df['City'].str.contains(city_name, case=False, na=False)]
    if not results.empty:
        return results['IATA'].tolist()
    return []

# Exemple d'utilisation
city = "New York"
iata_codes = get_all_iata(city, iata_filtered)
print(f"Les codes IATA pour {city} sont : {iata_codes}")
```

```python
# Fonction pour générer tous les trajets possibles en combinant les codes IATA
def generate_trajets(start_city, end_city, df):
    start_codes = get_all_iata(start_city, df)
    end_codes = get_all_iata(end_city, df)
    trajets = [f"{start} -> {end}" for start in start_codes for end in end_codes]
    return trajets

# Exemple d'utilisation
start_city = "New York"
end_city = "Paris"
all_trajets = generate_trajets(start_city, end_city, iata_filtered)
print(f"Tous les trajets possibles entre {start_city} et {end_city} : {all_trajets}")
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import time

# Récupérer les villes uniques
villes_uniques = pd.concat([df['start'], df['end']]).drop_duplicates().dropna().tolist()

# Initialiser Nominatim
geolocator = Nominatim(user_agent="my_app_name")

def get_coordinates_nominatim(city_name):
    try:
        location = geolocator.geocode(city_name, timeout=10)
        if location:
            return (location.latitude, location.longitude)
    except Exception as e:
        print(f"Erreur pour {city_name}: {e}")
    return None

coordonnees_villes = {}
for ville in villes_uniques:
    coordonnees = get_coordinates_nominatim(ville)
    coordonnees_villes[ville] = coordonnees
    print(f"{ville} -> {coordonnees}")
    time.sleep(1)

coord_df = pd.DataFrame.from_dict(coordonnees_villes, orient='index', columns=['latitude', 'longitude'])
coord_df.reset_index(inplace=True)
coord_df.rename(columns={'index': 'ville'}, inplace=True)
coord_df.to_csv("coordonnees_villes_nominatim.csv", index=False)

print("Coordonnées sauvegardées dans 'coordonnees_villes_nominatim.csv'.")
```

```python
from geopy.distance import geodesic

# Fonction pour estimer la distance et le temps de vol
def estimate_flight_time(lat1, lon1, lat2, lon2, avg_speed=900):
    """
    Calcule la distance géodésique entre deux points et estime le temps de vol.
    :param lat1: Latitude du point de départ
    :param lon1: Longitude du point de départ
    :param lat2: Latitude du point d'arrivée
    :param lon2: Longitude du point d'arrivée
    :param avg_speed: Vitesse moyenne de l'avion (en km/h)
    :return: Un dictionnaire avec la distance (km) et le temps de vol (heures)
    """
    coords1 = (lat1, lon1)
    coords2 = (lat2, lon2)
    distance = geodesic(coords1, coords2).kilometers
    flight_time = distance / avg_speed  # Temps estimé en heures
    return {"distance": distance, "flight_time": flight_time}
```

```python
from geopy.distance import geodesic

# Fonction pour vérifier si les coordonnées sont valides
def are_coordinates_valid(coords):
    return coords is not None and all(isinstance(value, (int, float)) and not pd.isna(value) for value in coords)

# Fonction pour calculer le temps de vol
def estimate_flight_time(lat1, lon1, lat2, lon2, avg_speed=900):
    if not are_coordinates_valid((lat1, lon1)) or not are_coordinates_valid((lat2, lon2)):
        return {"distance": None, "flight_time": None}
    coords1 = (lat1, lon1)
    coords2 = (lat2, lon2)
    distance = geodesic(coords1, coords2).kilometers
    flight_time = distance / avg_speed
    return {"distance": distance, "flight_time": flight_time}
```

```python
coord_dict = coord_df.set_index("ville")[["latitude", "longitude"]].T.to_dict()

# Fonction pour enrichir les trajets
def enrich_trajet(row):
    start_coords = coord_dict.get(row["start"])
    end_coords = coord_dict.get(row["end"])
    if start_coords and end_coords and are_coordinates_valid(start_coords.values()) and are_coordinates_valid(end_coords.values()):
        return estimate_flight_time(start_coords["latitude"], start_coords["longitude"],
                                    end_coords["latitude"], end_coords["longitude"])
    return {"distance": None, "flight_time": None}

# Appliquer la fonction
trajet_data = df.apply(enrich_trajet, axis=1)
trajet_df = pd.concat([df, trajet_data.apply(pd.Series)], axis=1)

# Sauvegarder les résultats
trajet_df.to_csv("trajets_with_flight_time.csv", index=False)
print("Dataset enrichi avec distances et temps de vol sauvegardé.")
```

```python
import requests

# Fonction pour récupérer le prix le plus bas pour un trajet unique via l'API Amadeus
def get_lowest_price(origin, destination, token, date="2025-01-30"):
    url = "https://test.api.amadeus.com/v2/shopping/flight-offers"
    params = {
        'originLocationCode': origin,
        'destinationLocationCode': destination,
        'departureDate': date,
        'adults': 1,
        'currencyCode': 'EUR',
        'includedAirlineCodes': 'AF,KL'
    }
    headers = {"Authorization": f"Bearer {token}"}
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()
        if 'data' in data and len(data['data']) > 0:
            prices = [float(offer['price']['total']) for offer in data['data']]
            lowest_price = min(prices)
            print(f"Prix le plus bas pour {origin} -> {destination}: {lowest_price} EUR (Air France/KLM)")
            return lowest_price
        else:
            print(f"Aucun vol trouvé pour {origin} -> {destination} (Air France/KLM).")
            return None
    else:
        print(f"Erreur API : {response.status_code} - {response.json()}")
        return None

def process_lowest_prices(trajets_df, token, date="2025-01-30"):
    results = []

    for _, row in trajets_df.iterrows():
        origin = row['start_iata']
        destination = row['end_iata']
        trajet = row['trajet']

        price = get_lowest_price(origin, destination, token, date)
        if price is not None:
            results.append({
                'trajet': trajet,
                'origin': origin,
                'destination': destination,
                'lowest_price': price
            })

    results_df = pd.DataFrame(results)
    return results_df

# Charger votre fichier CSV contenant les trajets
trajets_df = pd.read_csv("trajets_with_iata_and_comments.csv")


# Appliquer le traitement pour tous les trajets
results_df = process_lowest_prices(trajets_df, token)
results_df.to_csv("lowest_flight_prices_af_kl.csv", index=False)
print("Fichier 'lowest_flight_prices_af_kl.csv' créé avec les prix les plus bas pour Air France et KLM.")
```
